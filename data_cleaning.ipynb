#### We will seperately Clean all three tables that have been extracted and merge them together 

```python
mvps = pd.read_csv('Dataframes/mvp_votes.csv')
```

```python
mvps = mvps[["Player","Year","Pts Won", "Pts Max", "Share"]]
```

### Cleaning Player Data

```python
players = pd.read_csv("Dataframes/players.csv")
```

```python
players.isna().sum()
```

```python
del players['Unnamed: 0']
del players['Rk']
```

```python
players.head()
```

```python
#Removing rows with asterik on them to denote mvp winners
players["Player"] = players["Player"].str.replace("*","", regex=False)
```

Certain players played for multiple teams in a single year, we will merge their overall performance for that year into a single row

```python
players.groupby(["Player", "Year"])
```

```python
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(20)
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
df = players.merge(mvps, how = "outer", on =["Player", "Year"])
```

```python
df
```

```python
df[df["Pts Won"]  == 0]
```

```python
df[['Pts Won', 'Pts Max', 'Share']] = df[['Pts Won', 'Pts Max', 'Share']].fillna(0)
```

```python
df.tail()
```

### Cleaning Team Data

```python
teams = pd.read_csv('Dataframes/teams_standings.csv')
```

```python
teams.head(30)
```

```python
teams = teams[~teams['W'].str.contains("Division")]
```

```python
teams.head(30)
```

```python
teams['Team'] = teams['Team'].str.replace('*',"",regex=False)
```

```python
del teams['Unnamed: 0']
```

```python
teams.head(30)
```

```python
nicknames = {}

with open("Dataframes/nicknames.txt") as f:
    lines = f.readlines()
    for line in lines[1:]:
       abbrev, name =  line.replace("\n" , "").split(",")
       nicknames[abbrev] = name
```

```python
nicknames
```

```python
df['Team'] = df['Tm'].map(nicknames)
```

```python
df.head(10)
```

```python
teams['Year'] = teams['Years']
```

```python
del teams['Years']
teams.head()
```

```python
nba_stats = df.merge(teams, how='outer', on=['Team', 'Year'])
```

```python
nba_stats
```

```python
nba_stats = nba_stats.apply(pd.to_numeric, errors='ignore')

nba_stats.dtypes
```

```python
nba_stats["GB"] = pd.to_numeric(nba_stats["GB"].str.replace("—","0"))
```

```python
nba_stats.dtypes
```

```python
nba_stats.tail()
```

```python
nba_stats.isna().sum()
```

```python
nba_stats.to_csv("mvp_stats_df1.csv")
```
