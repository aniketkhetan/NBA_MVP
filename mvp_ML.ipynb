```python
#Defining the feature set
predictors = ['Age', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P',
       '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'Year', 
       'W', 'L', 'W/L%', 'GB', 'PS/G',
       'PA/G', 'SRS']

# predictors = ['Age', 'G', 'GS', 'MP',  'FG%',  '3P%',  '2P%',  'FT%', 'ORB',
#        'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'Year',
#        'W', 'L', 'W/L%', 'GB', 'PS/G',
#        'PA/G', 'SRS']
```

```python
print(f'Predictors = {predictors}')
```

```python
#Splitting training and test data
train = stats[(stats["Year"] < 2023)]
test = stats[stats["Year"] == 2023]
```

### Ridge Regression for only the year 2023

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=0.1)
```

```python
reg.fit(train[predictors],train['Share'])
```

```python
predictions = reg.predict(test[predictors])
```

```python
predictions
```

```python
#Visualising the predictions in a dataframe
#Here the predictions are sorted by Share and the top 10 are displayed
#By arranging the Rank and Pred_Rank columns, we can see how the predictions compare to the actual Share
def add_ranks(predictions):
    predictions = predictions.sort_values("predictions", ascending=False)
    predictions["Predicted_Rk"] = list(range(1,predictions.shape[0]+1))
    predictions = predictions.sort_values("Share", ascending=False)
    predictions["Rk"] = list(range(1,predictions.shape[0]+1))
    predictions["Diff"] = (predictions["Rk"] - predictions["Predicted_Rk"])
    return predictions
```

```python
#Predictions for the year 2023
predictions = pd.DataFrame(predictions,columns=['predictions'],index=test.index)
predictions_combined = pd.concat([test[['Player', 'Share','Year']],predictions],axis=1)
predictions_combined = add_ranks(predictions_combined)
predictions_combined.head(10)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error_ridge = mean_squared_error(predictions_combined["Share"],predictions_combined['predictions'])
mean_squared_error_ridge
```

```python
#Creating a function to calculate the error based on the top 5 predictions and the model's ability to predict them accurately
#Read attached paper for more details on how the 'AP' is calculated
def find_ap(combination):
    actual = combination.sort_values("Share", ascending=False).head(5)
    predicted = combination.sort_values("predictions", ascending=False)
    ps = []
    found = 0
    seen = 1
    for index,row in predicted.iterrows():
        if row["Player"] in actual["Player"].values:
            found += 1
            ps.append(found / seen)
        seen += 1

    return sum(ps) / len(ps)
```

```python
#For year 2023
AP = find_ap(predictions_combined)
AP
```

### Backtesting
Backtesting will allow us to run our models on a set of years and return an average performance across the inputted years.


```python
#Crreating a function to backtest the model
aps = []
all_predictions = []
years = list(range(1991,2024))
def backtesting(stats, model, years, predictors) :
    for year in years:
        
        train = stats[stats['Year'] < year]
        test = stats[stats['Year'] == year]
        model.fit(train[predictors],train['Share'])
        predictions = model.predict(test[predictors])
        predictions = pd.DataFrame(predictions, columns=['predictions'], index=test.index)
        predictions_combined = pd.concat([test[["Player", "Share",'Year']], predictions], axis=1)
        predictions_combined = add_ranks(predictions_combined)
        all_predictions.append(predictions_combined)
        aps.append(find_ap(predictions_combined))
    return sum(aps)/len(aps), aps ,  pd.concat(all_predictions),all_predictions
```

```python
#Backtesting the model to predict the results for the years 2001-2023
mean_aps_reg, aps, all_predictions_df,all_predictions = backtesting(stats,reg, years[10:],predictors)
```
